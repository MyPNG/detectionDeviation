# Single REA Step2-4 Pipeline Test

This notebook runs step 2 to step 4 for one reranked REA unit JSON (e.g. `rea-01#4.json`).

- Step 2: extract top-k main REG matches
- Step 3: collect one-hop REG graph context
- Step 4: generate prompt payload

```python
from pathlib import Path
import json
import sys

PROJECT_ROOT = Path('/Users/my/Documents/projects/detectionDeviation').resolve()
REASONING_DIR = PROJECT_ROOT / 'pipeline' / 'reasoning'
if str(REASONING_DIR) not in sys.path:
    sys.path.insert(0, str(REASONING_DIR))

from SingleReaStep2To4Pipeline import SingleReaStep2To4Pipeline
```

```python
# Input: one REA split-unit reranked file
reranked_json = PROJECT_ROOT / 'intermediate_outputs' / 'artifact_01_case3_v3_reranked' / 'chunk1_deontic_stages' / 'rea-01#4.json'
top_k = 5

output_dir = PROJECT_ROOT / 'intermediate_outputs' / 'single_rea_step2_4' / 'rea-01#4'

pipeline = SingleReaStep2To4Pipeline(project_root=PROJECT_ROOT)
result = pipeline.run_single(
    reranked_json_path=reranked_json,
    output_dir=output_dir,
    top_k=top_k,
)
print(json.dumps(result, indent=2, ensure_ascii=False))
```

```python
step2 = json.loads((output_dir / 'step2_main_reg_matches.json').read_text(encoding='utf-8'))
step3 = json.loads((output_dir / 'step3_graph_context.json').read_text(encoding='utf-8'))
step4 = json.loads((output_dir / 'step4_prompt_payload.json').read_text(encoding='utf-8'))

print('Step2 main REG IDs:', [row['reg_id'] for row in step2.get('main_reg_matches', [])])
print('Step3 graph neighbor count:', len(step3.get('one_hop_neighbors', [])))
print('\nPrompt preview:\n')
print(step4.get('user_prompt', '')[:2500])
```
